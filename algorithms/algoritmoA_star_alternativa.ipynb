```python
grafo = {}
grafo['S0']= [['S1',6], ['S2', 4]]
grafo['S1']= [['S3',1]]
grafo['S2']= [['S3',2]]
grafo['S3']= [['S4',4], ['S1', 5], ['S5',6]]
grafo['S4']= [['S5',3], ['S3', 4]]
grafo['S5']= [['S1',5]]
```

```python
h = {}
h['S0'] = 10
h['S1'] = 5
h['S2'] = 8
h['S3'] = 4
h['S4'] = 3
h['S5'] = 0
```

```python
class Node:
    def __init__(self, state, parent, partial_path, f):
        self.state=state
        self.parent=parent
        self.partial_path=partial_path
        self.heuristic=f

    def print_path(self):
        if self.parent is not None:
            self.parent.print_path()
        print("->", self.state.name)
            
    
    def __lt__(self,other):
        return self.heuristic<other.heuristic
        
```

```python
class State:
    def __init__(self, name=None):
        if name is None:
            self.name= self.getInitialState()
        else:
            self.name=name

    def getInitialState(self):
        return 'S0'

    def successorFunction(self):
        return grafo[self.name]
        
    def checkGoalState(self):
        return self.name=='S5'
```

```python
import queue
```

```python
def A_TreeSearch():
    fringe=queue.PriorityQueue()

    initialState= State()
    euristica=h[initialState.name]
    root=Node(initialState, None, 0, euristica)
    
    fringe.put((root.heuristic, root))

    while not fringe.empty():
        _, currentNode= fringe.get()
        print("dequeue", currentNode.state.name)
        if currentNode.state.checkGoalState():
            print("Soluzione trovata")
            currentNode.print_path()
            break

        childStates= currentNode.state.successorFunction()
        for (childState, distance) in childStates:
            g=distance + currentNode.partial_path
            euristica=h[childState]
            f= g + euristica
            childNode= Node(State(childState), currentNode, g, f)
            fringe.put((childNode.heuristic, childNode))
```

```python
A_TreeSearch()
```
