```python
class State:
    def __init__(self, name= None):
        if name==None:
            self.name=self.getInitialState()  #crea lo stato di partenza
        else:
            self.name= name

    def getInitialState(self):
        initialState='S0'
        return initialState

    def successorFunction(self):
        return grafo[self.name]  #ritorna i vicini definiti del grafo

    def checkGoalState(self):
        return self.name=='S5'
```

```python
import queue as Queue
```

```python
def A_Star_TreeSearch():
    fringe=Queue.PriorityQueue()  #crea una coda di priorità che serve come frontiera (lista aperta) per l'algoritmo
    # crea il root node
    initialState= State()
    euristica= h[initialState.name]  #Ottiene il valore euristico per questo stato
    root= Node(initialState, None, euristica, 0)    #il padre della radice é None

    fringe.put((root.heuristic, root))  #Aggiunge il nodo radice alla coda di priorità, utilizzando il suo valore euristico come priorità.

    while not fringe.empty():
        _, currentNode= fringe.get()  #estrae il front node dalla frontiera

        print("-- dequeue --", currentNode.state.name)   #stampa il nodo appena estratto

        #verifica se é lo stato obiettivo
        if currentNode.state.checkGoalState():
            print("Stato obiettivo raggiunto")
            print("Soluzione")
            currentNode.printPath()
            break

        childStates = currentNode.state.successorFunction()
        for(childState, distance) in childStates: 
            g= currentNode.partial_path + distance   #Per ogni successore, calcola il nuovo costo del percorso (g)
            euristica= h[childState]   # Ottiene il valore euristico per il successore
            f=g+ euristica
            childNode= Node(State(childState), currentNode, f, g)   #Crea un nuovo nodo per il successore

            fringe.put((childNode.heuristic, childNode))   #Aggiunge il nuovo nodo alla coda di priorità
        
```

```python
def A_Star_GraphSearch():

    fringe= Queue.PriorityQueue()   #Crea una coda di priorità per memorizzare i nodi da esplorare, ordinati in base ai loro valori euristici.
    close=[]    #lista vuota per tenere traccia degli stati visitati

    initialState= State()
    euristica= h[initialState.name]
    root= Node(initialState, None, euristica,0)  # nodo radice con lo stato iniziale, nessun genitore, il valore euristico e un costo del percorso di 0

    fringe.put((root.heuristic, root))  #Aggiunge il nodo radice alla coda di priorità con la sua euristica come priorità.

    while not fringe.empty():
        _, currentNode= fringe.get()   #Estrae il nodo con il valore euristico più basso
        print("--dequeue--", currentNode.state.name)

        if currentNode.state.checkGoalState():
            print("Stato obiettivo raggiunto")
            print("Soluzione:")
            currentNode.printPath()
            break

        if (currentNode.state.name) not in close:
            close.append(currentNode.state.name)
            
            childStates= currentNode.state.successorFunction()    # aggiunge lo stato alla lista dei visitati e genera stati successori
            for(childState, distance) in childStates:
                g=currentNode.partial_path+ distance
                euristica= h[childState]
                f=g+euristica
                childNode= Node(State(childState), currentNode, f, g)

                if childNode.state.name not in close:
                    fringe.put((childNode.heuristic, childNode))
```
