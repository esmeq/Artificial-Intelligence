```python
h = {}
h['Arad'] = 366
h['Bucarest'] = 0
h['Craiova'] = 160
h['Drobeta'] = 242
h['Eforie'] = 161
h['Fagaras'] = 176
h['Giurgiu'] = 77
h['Hirsova'] = 151
h['Iasi'] = 226
h['Lugoj'] = 244
h['Mehadia'] = 241
h['Neamt'] = 234
h['Oradea'] = 380
h['Pitesti'] = 100
h['Rimnicu Vilcea'] = 193
h['Sibiu'] = 253
h['Timisoara'] = 329
h['Urziceni'] = 80
h['Vaslui'] = 199
h['Zerind'] = 374
```

```python
class Node:
    def __init__(self, state, parent, h):
        self.state=state
        self.parent=parent
        self.heuristic=h

    def printPath(self):
        if self.parent is not None:
           self.parent.printPath()
        print("--> ", self.state.name)

    def __lt__(self, other):
        return self.heuristic<other.heuristic
```

```python
class State:
    def __init__(self, name=None):
        if name is None:
            self.name= self.getInitialState()
        else:
            self.name=name
            
    def getInitialState(self):
        return'Arad'
        
    def successorFunction(self):
        return connections[self.name]
    
    def checkGoalState(self):
        return self.name == 'Bucarest'
```

```python
import queue as Queue
```

```python
def A_Greedy():
    fringe= Queue.PriorityQueue()

    initialState= State("Arad")
    euristica=h[initialState.name]
    root=Node(initialState, None, euristica)
    fringe.put((root.heuristic, root))
    
    while not fringe.empty():
        _, currentNode= fringe.get()
        print("-- dequeue --", currentNode.state.name)
        
        if currentNode.state.checkGoalState():
            print("Soluzione trovata")
            currentNode.printPath()
            break

        childStates=currentNode.state.successorFunction()
        for childState in childStates:
            euristica=h[childState]
            childNode=Node(State(childState), currentNode, euristica)

            fringe.put((childNode.heuristic, childNode))
```

```python
A_Greedy()
```
