```python
scacchiera=[0,1,2,3,4]


def calcolaConflitti(scacchiera):
    conflitti=0
    for i in range(len(scacchiera)-1):
        for j in range(i+1,len(scacchiera)):
            if scacchiera[i]==scacchiera[j]:
                conflitti+=1
            elif abs(scacchiera[i]-scacchiera[j]) == abs(i-j):
                conflitti+=1
    return conflitti

def generaVicini(scacchiera):
    vicini=[]
    for colonna in range(len(scacchiera)):
        for riga in range(len(scacchiera)):
            if riga is not scacchiera[colonna]:    #se la regina si  è spostata e non sta nella stessa riga
                copiaScacchiera=list(scacchiera)
                copiaScacchiera[colonna]=riga   #sposto regina
                costo= calcolaConflitti(copiaScacchiera)
                vicini.append((costo,copiaScacchiera))   #tupla con mossa e suo costo             
    return vicini   
```

```python
def TabuSearch5Regine(scacchiera):
    migliorScacchiera= scacchiera
    migliorCosto= calcolaConflitti(scacchiera)
    print(f"Scacchiera corrente {scacchiera} con conflitti {migliorCosto}")

    currentScacchiera=scacchiera
    currentCosto= calcolaConflitti(scacchiera)
    
    tabuList=[]
    mosseTabu=5
    iterazione=0
    max_iterazioni=1000
    while currentCosto>0 and iterazione<max_iterazioni:   #a zero ho trovato la soluzione
        vicini= generaVicini(currentScacchiera)  #genera tutte le possibili mosse nella scacchiera attuale
        vicini.sort()    #riordina tutti i vicini dal piu piccolo

        migliormossa=None
        for costoVicino, scacchieraVicino in vicini:
            if scacchieraVicino not in tabuList or costoVicino<migliorCosto:     #controlla gia in ordine dalla mossa migliore grazie a sort di prima
                currentScacchiera= scacchieraVicino
                currentCosto= costoVicino
                migliormossa= True
                break
        if not migliormossa:
            indiceVecchio= float('inf')
            for costoVicino, scacchieraVicino in vicini:
                indiceCorrente= tabuList.index(scacchieraVicino)
                if indiceCorrente<indiceVecchio:
                    indiceVecchio=indiceCorrente
                    currentScacchiera= scacchieraVicino
                    currentCosto= costoVicino
             if indiceVecchio is not float('inf'):
                tabuList.pop(indiceVecchio)

        if currentCosto<migliorCosto:
            migliorScacchiera=currentScacchiera
            migliorCosto=currentCosto

        tabuList.append(currentScacchiera)
        if len(tabuList)>mosseTabu:
            tabuList.pop(0)

        iterazione+=1

    print(f"Miglior scacchiera {migliorScacchiera} con costo {migliorCosto} dopo {iterazione} iterazioni") 
```

```python
TabuSearch5Regine(scacchiera)
```
