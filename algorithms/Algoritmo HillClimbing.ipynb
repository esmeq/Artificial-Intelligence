```python
import random
```

```python
scacchiera = [random.randint(0,7) for _ in range (8)]
#calcolo dei conflitti
def contaConflitti(scacchiera):
    conflitti=0
    for i in range (len(scacchiera)-1):    #controlla se ci sono conflitti nella stessa riga
        for j in range(i+1, len(scacchiera)):
            if scacchiera[i]==scacchiera[j]:
                conflitti+=1
            elif abs(scacchiera[i]-scacchiera[j])==abs(i-j):   #riga i-riga j == colonna i-colonna j
                    conflitti +=1
    return conflitti

#cerco miglior vicino
def migliorVicino(scacchiera):
    migliorCosto= contaConflitti(scacchiera) 
    migliorScacchiera= scacchiera
    for colonna in range (8):
        for riga in range(8):
            if scacchiera[colonna]==riga:   #la reginga non si è spostata
                continue   #salta  queste righe sotto e va avanti con l'indice
            vicino=list(scacchiera)  #creo una copia di scacchiera
            vicino[colonna]=riga
            conflittiVicino= contaConflitti(vicino)
            if conflittiVicino<migliorCosto:
                migliorCosto=conflittiVicino
                migliorScacchiera= vicino
                
    return migliorCosto, migliorScacchiera
```

```python
def HillClimbingClassico():
    scacchieraCorrente = [random.randint(0,7) for _ in range (8)]
    costoCorrente= contaConflitti(scacchieraCorrente)
    print(f"Punto di partenza {scacchieraCorrente}, costo {costoCorrente} ")
    mosse=0
    while True:
        costoVicino, prossimaScacchiera= migliorVicino(scacchieraCorrente)
        if costoVicino<costoCorrente:
            scacchieraCorrente=prossimaScacchiera
            costoCorrente=costoVicino
            mosse+=1
            print(f"Mosse fatte: {mosse}, scacchiera {scacchieraCorrente}, costo: {costoCorrente} ")
        else:
            break
    print("Risultato finale")
    if costoCorrente==0:   #conflitti a zero, trovato minimo locale
        print(f"Problema risolto in mosse: {mosse} ")
        print(f"Scacchiera: {scacchieraCorrente},") 
    else:
        print("L'algoritmo si è fermato in minimo locale")
        print(f"Scacchiera: {scacchieraCorrente},")
```

```python
grafo = {}
grafo['S0']= ['S1','S2']
grafo['S1']= ['S3']
grafo['S2']= ['S3']
grafo['S3']= ['S4', 'S1', 'S5']
grafo['S4']= ['S5', 'S3']
grafo['S5']= ['S1']
```

```python
h = {}
h['S0'] = 10
h['S1'] = 5
h['S2'] = 8
h['S3'] = 4
h['S4'] = 3
h['S5'] = 0
```

```python
class Node:
    def __init__(self, state, parent, h):
        self.state=state
        self.parent=parent
        self.heuristic=h
        
    def print_path(self):
        if self.parent is not None:
            self.parent.print_path()
        print("->", self.state.name)
    
```

```python
class State:
    def __init__(self, name=None):
        if name is None:
            self.name= self.getInitialState()
        else: 
            self.name=name

    def getInitialState(self):
        return "S0"

    def successorFunction(self):
        return grafo[self.name]
        
    def checkGoalState(self):
        return self.name=="S5"
```

```python
def HillClimbingClassico():
    initialState= State()
    euristica=h[initialState.name]
    currentNode=Node(initialState, None, euristica)

    while True:
        controllo=currentNode
        if currentNode.state.checkGoalState():
            print("Soluzione trovata")
            currentNode.print_path()
            break
        bestChild=None
        bestHeuristic=currentNode.heuristic
        childStates=currentNode.state.successorFunction()
        for childState in childStates:
            nuovostato=State(childState)
            euristica=h[nuovostato.name]
            childNode=Node(nuovostato, currentNode, euristica)
            if childNode.heuristic<bestHeuristic:
                bestHeuristic=childNode.heuristic
                bestChild=childNode

        if bestChild is not None:
            currentNode=bestChild

        if currentNode is controllo:
            print("L'algoritmo si é incastrato in un minimo locale")
            currentNode.print_path()
            break              
```

```python
def RandomRestartHillClimbing(max_tentativi=100):
    for i in range(max_tentativi):   
        initialState= State()
        euristica=h[initialState.name]
        currentNode=Node(initialState, None, euristica)

        while True:
            controllo=currentNode
            if currentNode.state.checkGoalState():
                print("Soluzione trovata")
                currentNode.print_path()
                return
            bestChild=None
            bestHeuristic=currentNode.heuristic
            childStates=currentNode.state.successorFunction()
            for childState in childStates:
                nuovostato=State(childState)
                euristica=h[nuovostato.name]
                childNode=Node(nuovostato, currentNode, euristica)
                if childNode.heuristic<bestHeuristic:
                    bestHeuristic=childNode.heuristic
                    bestChild=childNode

            if bestChild is not None:
                currentNode=bestChild

            if currentNode is controllo:
                print(f"Tentativo {i+1}, l'algoritmo si è incastrato in un minimo locale")
                currentNode.print_path()
                break              
```

```python
#random restartHillClimbing
import random
import string

lista= ["A", "B", "%", "Z", "X", "/"]
max_tentativi=20
target= "".join(random.choices(string.printable, k=max_tentativi))
tentativi=10
def CalcolaDistanza(stringa,target):
    distanza=0
    for i in range(len(stringa)):
        s=stringa[i]
        t=target[i]
        delta=abs(ord(s)-ord(t))
        distanza=distanza+delta
    return distanza

def tweak(stringa):
    indice= random.randint(0, len(stringa)-1)
    stringa[indice]=random.choice(lista)

def RicercaStringa(target, max_tentativi, tentativi):

    for i in range(max_tentativi):
        currentState= "".join(random.choices(string.printable, k=max_tentativi))
        currentDistance= CalcolaDistanza(currentState,target)
        mosse=0
        print(f"--RESTART {i+1} --")
        while True:
            controllo=currentState
            if currentDistance == 0:
                print(f" Successo, numero tentativi {mosse}")
                print("Stringa finale: ", ''.join(currentState))
                return
            j=0
            while j<tentativi:
                nextState= list(currentState)
                tweak(nextState)

                nextDistance= CalcolaDistanza(nextState,target)
                if nextDistance<currentDistance:
                    currentState= nextState
                    currentDistance=nextDistance
                    mosse+=1
                    print(f"Distanza {currentDistance} , stringa {''.join(currentState)}")
                    break
            if currentState == controllo:
                break
                
    print("Soluzione non trovata")            
```

```python
RicercaStringa(target,max_tentativi, tentativi)
```
