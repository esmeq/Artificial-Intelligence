```python
import math
import random
A=(0,0)
B=(2,8)
C=(5,2)
D=(7,9)
E=(9,3)

nomeCittà = {
    (0, 0): 'A',
    (2, 8): 'B',
    (5, 2): 'C',
    (7, 9): 'D',
    (9, 3): 'E'
}

lista=[A,B,C,D,E]
def distanzaTotale(lista):
    distanza=0
    distanzaTot=0
    for i in range(len(lista)-1):     
        x1,y1= lista[i]
        x2,y2= lista[i+1]
        distanza= math.sqrt((x2-x1)**2 + (y2-y1)**2)
        distanzaTot+=distanza
    return distanzaTot

def scambiaCittà(percorso):
    nuovoPercorso= list(percorso)
    indice1=random.randint(0,len(nuovoPercorso)-1)
    indice2=random.randint(0,len(nuovoPercorso)-1)
    while indice1 == indice2:
        indice2= random.randint(0,len(nuovoPercorso)-1)

    nuovoPercorso[indice1],nuovoPercorso[indice2]= nuovoPercorso[indice2],nuovoPercorso[indice1]    #scambia città
    return nuovoPercorso
    
def SimulatedAnnealingDrone():
    currentState= lista
    currentDistance= distanzaTotale(currentState)
    T=100
    alpha=0.95   #coefficiente raffreddamento   vicino all'1 si abbassa lentamente ma è piu preciso
    migliorPercorso= currentState
    migliorDistanza= currentDistance
    while T>0.01 :
        print(f"Percorso: {currentState}, distanza: {currentDistance}, temperatura: {T} ")
        nextState= scambiaCittà(currentState)   #genera vicino
        nextDistance=distanzaTotale(nextState)
        delta_e= nextDistance-currentDistance

        if delta_e<0:
            currentState= nextState
            currentDistance= nextDistance
            
        else:
            prob=math.exp(-delta_e/T)
            if random.random()<prob:    #random.random da un numero tra 0.0 e 0.99
                currentState= nextState
                currentDistance= nextDistance
                print("Passo peggiorativo accettato con probabilità: ", prob)
        
        if currentDistance<migliorDistanza:
            migliorPercorso= currentState
            migliorDistanza= currentDistance
        
        T=T*alpha   #abbasso temperatura
        
    nomipercorso= [nomeCittà[punto] for punto in migliorPercorso]
    print("Miglior percorso trovato: ")
    print(" -> ".join(nomipercorso))
```

```python
SimulatedAnnealingDrone()
```
