```python
import numpy as np
```

## Funzione di Ackley

```python
def ackley(x, y):
    term1 = -20.0 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2)))
    term2 = -np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y)))
    return term1 + term2 + np.e + 20
```

## Algoritmo Simulated Annealing

```python
def simulated_annealing(
        func,
        bounds=[(-5, 5), (-5, 5)],
        T0=10.0,
        Tmin=1e-3,
        alpha=0.95,
        L=100,                 # iterazioni per temperatura
        step_size=0.5
    ):
    
    # Punto iniziale casuale nel range
    x = np.random.uniform(bounds[0][0], bounds[0][1])
    # INSERISCI QUI IL TUO CODICE
    y = np.random.uniform(bounds[1][0], bounds[1][1])
    
    current_value = func(x, y)
    best_x, best_y = x, y
    best_value = current_value
    
    T = T0
    
    # Loop sulle temperature
    while T > Tmin:
        
        # Loop interno: L iterazioni a temperatura fissa
        for _ in range(L):
            
            # Generazione stato next (perturbazione gaussiana)
            x_new = x + np.random.normal(0, step_size)
            y_new = y + np.random.normal(0, step_size)
            
            # Rispetta i bounds
            x_new = np.clip(x_new, bounds[0][0], bounds[0][1])
            y_new = np.clip(y_new, bounds[1][0], bounds[1][1])
            
            # INSERISCI QUI IL TUO CODICE
            new_value = func(x_new,y_new)
            delta = new_value - current_value
            
            # Criterio di accettazione (Metropolis)
            if delta < 0 or np.random.rand() < np.exp(-delta / T):
                # INSERISCI QUI IL TUO CODICE
                x=x_new
                y=y_new
                current_value=new_value
                
                
                # Aggiorna migliore soluzione
                if new_value < best_value:
                    best_x, best_y = x_new, y_new
                    best_value = new_value
        
        # Raffreddamento
        T = alpha * T
    
    return best_x, best_y, best_value
```

## Esecuzione

```python
best_x, best_y, best_val = simulated_annealing(ackley)
    
print('Miglior soluzione trovata:')
print('x =', best_x)
print('y =', best_y)
print('f(x,y) =', best_val)
```
